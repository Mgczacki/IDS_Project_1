```python
siblings_ratings = data_df[data_df['Are you an only child? (1: Yes; 0: No; -1: Did not respond)'] == 0].iloc[:,0:400]
siblings_ratings
```

```python
siblings_ratings.reset_index(inplace = True)
```

```python
siblings_ratings
```

```python
# 0-176 are only child rates, and 177-1070 are the siblings rates
df_append = only_child_ratings._append(siblings_ratings, ignore_index=True)
```

```python
df_append
```

```python
df_append = df_append.drop('index', axis=1)
```

```python
df_append
```

```python
count = 0
for column in df_append:
     
    # Select column contents by column
    # name using [] operator
    columnSeriesObj = df_append[column]
    only_child = columnSeriesObj[:177].dropna()
    siblings = columnSeriesObj[177:].dropna()
    res = ttest_ind(only_child, siblings)
    if res[1] <= 0.005:
        count += 1

    print('Column Name : ', column)
    print('Column Contents : ', columnSeriesObj.values)
    print('ratings for only child : ', only_child.values)
    print('ratings for siblings : ', siblings.values)
    print('results : ', res)
    print('Count for p value smaller than 0.005 : ', count)
```

```python
M1 = pd.to_numeric(data_df['Star Wars: Episode IV - A New Hope (1977)'],errors='coerce').values
M2 = pd.to_numeric(data_df['Star Wars: Episode II - Attack of the Clones (2002)'],errors='coerce').values
M3 = pd.to_numeric(data_df['Star Wars: Episode V - The Empire Strikes Back (1980)'],errors='coerce').values
M4 = pd.to_numeric(data_df['Star Wars: Episode 1 - The Phantom Menace (1999)'],errors='coerce').values
M5 = pd.to_numeric(data_df['Star Wars: Episode VII - The Force Awakens (2015)'],errors='coerce').values
M6 = pd.to_numeric(data_df['Star Wars: Episode VI - The Return of the Jedi (1983)'],errors='coerce').values
```

```python
M1 = M1[np.isfinite(M1)] # only keep the finite elements (not infinity or NaN)
M2 = M2[np.isfinite(M2)] # only keep the finite elements (not infinity or NaN)
M3 = M3[np.isfinite(M3)] # only keep the finite elements (not infinity or NaN)
M4 = M4[np.isfinite(M4)] # only keep the finite elements (not infinity or NaN)
M5 = M5[np.isfinite(M5)] # only keep the finite elements (not infinity or NaN)
M6 = M6[np.isfinite(M6)] # only keep the finite elements (not infinity or NaN)
```

```python
#%% 3 Reformatting - V1: Format data into a representation to work with
combinedData = np.transpose(np.array([M1,M2,M3,M4,M5,M6], dtype=object)) # array of arrays
    
#%% 4a) Extrastriate cortex: Doing the actual specialized analyses
# Descriptive statistics - we are looking for very special numbers that
# capture the essence of the entire dataset - the typical number (central
# tendency) and the dispersion. Typically mean and SD. 
numMovies = 6
# Initialize container to store descriptives:
descriptivesContainer = np.empty([numMovies,4])
descriptivesContainer[:] = np.NaN 

# 1. Element-wise:
for ii in range(numMovies):
    descriptivesContainer[ii,0] = np.mean(combinedData[ii]) # mu
    descriptivesContainer[ii,1] = np.std(combinedData[ii]) # sigma
    descriptivesContainer[ii,2] = len(combinedData[ii]) # n
    descriptivesContainer[ii,3] = descriptivesContainer[ii,1]/np.sqrt(descriptivesContainer[ii,2]) # sem
```
